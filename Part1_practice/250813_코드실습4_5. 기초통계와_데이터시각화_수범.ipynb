```python
import numpy as np

x = [5, 7, 8, 9, 10, 12, 13, 14, 20 ]
q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)
iqr = q3 - q1

print("Q1: ", q1)
print("Q2 (중앙값): ", q2)
print("Q3: ", q3)
print("IQR: ", iqr)
```

```python
# 박스 플롯
plt.boxplot(x, vert=False) # vert가 기본으로 들어가서 90도로 돌려짐!!False를 사용해 돌리지 않는다고 설정함
plt.title("사분위수의 IQR 시각화 (Boxplot)")
plt.xlabel("값")
plt.show()
```

```python
x = [-100, 5, 7, 8, 9, 10, 12, 13, 14, 20, 79]
q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)
iqr = q3 - q1

print("Q1: ", q1)
print("Q2 (중앙값): ", q2)
print("Q3: ", q3)
print("IQR: ", iqr)
```

```python
plt.boxplot(x, vert=False)
plt.title("사분위수의 IQR 시각화 (Boxplot)")
plt.xlabel("값")
plt.show()
```

```python
# 박스 플롯 - 커스터마이징

plt.boxplot(
    x,
    vert = False,
    patch_artist=True,                                        # 박스 색을 채워라
    boxprops=dict(facecolor = 'skyblue', color = 'blue'),     # 박스 스타일
    medianprops=dict(color = 'red', linewidth = 3),           # 중앙값 선 스타일
    whiskerprops=dict(color = 'gray', linestyle = '--'),      # 수염 스타일
    flierprops=dict(marker='*', markersize=8)                 # 이상치 스타일
    )
plt.title("사분위수 시각화", fontsize=15, fontweight='bold')
plt.grid(axis = 'x', linestyle = '--', alpha=0.4)
plt.yticks([1], ['Group A'])
plt.show()
```

```python
# 박스 플롯 - 커스터마이징(2)

box_style=dict(facecolor='skyblue', color='blue')              # 박스 스타일
median_style=dict(color='red', linewidth=3)                    # 중앙값 선 스타일
whisker_style=dict(color='gray', linestyle='--')               # 수염 스타일
flier_style=dict(marker='*', markersize=8)                     # 이상치 스타일

plt.boxplot(
    x,
    vert = False,
    patch_artist=True,                                         # 박스 색을 채워라
    boxprops=box_style,                                        # 박스 스타일
    medianprops=median_style,                                  # 중앙값 선 스타일
    whiskerprops=whisker_style,                                # 수염 스타일
    flierprops=flier_style                                     # 이상치 스타일
    )
plt.title("사분위수 시각화", fontsize=15, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.yticks([1], ['Group A'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.boxplot(
    df['height'],
    vert = False,
    patch_artist=True,
    boxprops=dict(facecolor = 'skyblue', color = 'blue'),
    medianprops=dict(color = 'red', linewidth = 5),
    whiskerprops=dict(color = 'gray', linestyle = '--'),
    flierprops=dict(marker='*', markersize=8)
    )
plt.title("Height")
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(
    df['weight'],
    vert = False,
    patch_artist=True,
    boxprops=dict(facecolor = 'skyblue', color = 'blue'),
    medianprops=dict(color = 'red', linewidth = 5),
    whiskerprops=dict(color = 'gray', linestyle = '--'),
    flierprops=dict(marker='*', markersize=8)
    )
plt.title("Weight")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Seaborn 실습

# 산점도

sns.scatterplot(data=df, x='height', y='weight')
plt.title("키 VS 몸무게 산점도")
plt.show()
```

```python
# 히스토그램 - 구간 나누기

df['height_bin'] = pd.cut(df['height'], bins=range(150, 200, 5))         # cut을 통해 (소괄호<< ]대괄호<<
df['weight_bin'] = pd.cut(df['weight'], bins=range(40, 120, 10))
df
```

```python
# barplot

sns.barplot(data=df, x='height_bin', y='weight')
plt.title("키와 몸무게 평균 비교")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(data=df, x='height_bin', y='weight')
plt.title("키와 몸무게 평균 비교")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(data=df, x='weight_bin', y='height')
plt.title("키와 몸무게 평균 비교")
plt.xticks(rotation=45)
plt.show()
```

```python
# 박스플롯 - 구간별로
plt.figure(figsize=(12, 5))
sns.boxplot(data=df, x='height_bin', y='weight')
plt.title("키 박스플롯")
plt.xticks(rotation=45)
plt.show()
```

```python
# 바이올린 플롯 - 박스플롯 + 데이터의 밀도 결합

plt.figure(figsize=(12, 5))
sns.violinplot(data=df, x='height_bin', y='weight')
plt.title("키 바이올린 플롯")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 스트립 플롯

plt.figure(figsize=(12, 5))
sns.stripplot(data=df, x='height_bin', y='weight', jitter=True)
plt.title("키 스트립 플롯")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 스왐플롯

plt.figure(figsize=(12, 5))
sns.swarmplot(data=df, x='height_bin', y='weight')
plt.title("키 스트립 플롯")
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# 히스토 그램

sns.histplot(data=df, x='height')
plt.title("키 히스토그램")
plt.show()
```

```python
# 조인트 플롯

sns.jointplot(data=df, x='height', y='weight', kind="hex")
plt.show()
```

```python
# 페어플롯

sns.pairplot(df)
plt.show()
```

# 상관관계

두변수끼리 연관관계가 있는지 확인

```python
corr = titanic_data.select_dtypes(include="number").corr()              # 수치형 데이터만 상관관계 계산 가능 ex) 남자, 여자 (X) 남자=0, 여자=1 (O)
                                                                        # corr함수로 구현

sns.heatmap(corr, cmap='coolwarm', annot=True)                          # coolwarm
plt.show()
```

```python
# @title 결측치를 히트맵으로 시각화

sns.heatmap(titanic_data.isnull(), cbar=False, yticklabels=False)               # isnull()<<결측치 함수
plt.show()                                                                      # 그래프를 보고 밝은 부분은 보고 데이터값이 빠진것을 확인한다.
```
