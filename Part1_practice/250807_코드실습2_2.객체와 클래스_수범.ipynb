```python
class BankAccount:
  def __init__(self, owner, balance=0):
    self.owner = owner
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    return f'{amount}원이 입금됨. 현재 잔액: {self.balance}원.'

  def withdraw(self, amount):
    if self.balance < amount:
      return '잔액이 부족하여 출금 할수 없음.'
    else :
      self.balance -= amount
    return f'{amount}원이 출금됨. 현재 잔액: {self.balance}원'

  def transfer(self, target_amount, amount):
    if self.balance < amount:
      return '잔액이 부족하여 이체 실패'
    self.balance -= amount
    target_amount.balance =+ amount
    return (f'{amount}원 {target_amount.owner}님께 이체 완료\n'
            f'내 잔액: {self.balance}원')
```

```python
a1 = BankAccount('홍길동')
print(a1.deposit(10000))
```

```python
a1.withdraw(3000)
```

```python
a2 = BankAccount('김철수')
print(a2.deposit(5000))
```

```python
a2.withdraw(30000)
```

```python
a1.transfer(a2, 3000)
```

```python
# 실습문제1 상품 클래스 구현하기 문제
class Product:
    def __init__(self, name, price, stock=0):
        #### 여기에 구현 ####
        self.name = name
        self.price = price
        self.stock = stock

    def restock(self, amount):
        #### 재고 추가 구현 ####
        self.stock += amount
        return (f'{amount}개가 재고에 추가됨. 현재 재고: {self.stock}개')

    def info(self):
        #### 상품 설명 구현 ####
        self.info = f'상품명: {self.name}, 가격: {self.price}, 재고: {self.stock}'
        return self.info

    def sell(self, quantity):
        #### 판매 정보 구현 ####
        if self.stock < quantity:
          return '재고 부족'
        self.stock -= quantity
        return f'{quantity}개가 판매되었습니다. 남은 재고: {self.stock}'
```

```python
x = Product('사과', 800)
y = Product('바나나', 2000)
```

```python
y.restock(20)
```

```python
print(y.info())
```

```python
print(y.sell(20))
```

```python
class Person:
  def __init__(self, name, age, email=None):
    if not (0<= age <= 120):
      raise ValueError("유효한 나이를 입력하세요")
    self.name = name
    self.age = age
    self.email = email or 'unknown@email.com'

  def greet(self):
    print(f"{self.name} ({self.age}세) - {self.email} 님, 안녕하세요!")
```

```python
p1 = Person('Min',-10)
```

```python
p2 = Person('June', 23, 'abcd@email.com')
print(p2.greet())
```

### 3. 다양한 메소드

```python
# __str__메소드

class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = year

  def __repr__(self):
    return f'Book(title={self.title}, author={self.author}, year={self.year})'

  def __str__(self):
    return f'{self.title} by {self.author} ({self.year})'
```

```python
book1 = Book("노인과 바다", "헤밍웨이", "1932")
book2 = Book("소년이 온다", "한강", "2014")
```

```python
print(book1)
```

```python
print(book2)
```
