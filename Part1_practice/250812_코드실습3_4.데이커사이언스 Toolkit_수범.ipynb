```python
# 산점도 - 두값의 관계를 보고 싶을 때

x = np.random.rand(100)
y = x + np.random.normal(0, 0.1, 100)

plt.scatter(x, y, alpha=0.6, color='teal', s=40)
plt.title('삼전도')
plt.xlabel('X')
plt.ylabel('Y')
plt.xticks(rotation=45)
plt.show()
#해당 그래프는 X가 상승하면서 Y도 상승하면 양의 상관관계라고 한다.
```

```python
# 히스토그램 - 일상적으로 연속적인 데이터를 시각화 할때 사용한다.

data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, density=True, color='orange', edgecolor="black")
# density=True 확율밀도 그래프로 변경/입력하지 않으면 갯수로 표시
plt.title('히스토그램')
plt.show()
```

```python
# 파이차트

labels = ['A', 'B', 'C', 'D'] # labels는 다른 단어로 작성 가능
sizes = [25, 30, 20, 25]

plt.pie(sizes, labels=labels)
plt.title('파이 차트')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 30, 20, 25]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0, 0.1, 0, 0)# 특정 값을 부각시키기 위해서

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',              # 파이 조각 위에 "퍼센트" 보여주기
    startangle=180,                  # 시작 각도 조절
    explode=explode,                # 특정 조각 강조
    shadow=True,                    # 그림자 효과
    counterclock=False              # 시계 방향으로 돌려 보기
)
plt.title('커스터 마이징된 파이트', fontsize=14, fontweight='bold')
plt.tight_layout()                  # 그래프 겹쳐보이기 금지
plt.show()
```

포맷 문자열	설명	예시 출력
%1.0f%%	정수로 표시	25%
%1.1f%%	소수점 한 자리	25.0%
%1.2f%%	소수점 두 자리	25.00%


| 포맷 문자열    | 설명       | 예시 출력    |
| --------- | -------- | -------- |
| `%1.0f%%` | 정수로 표시   | `25%`    |
| `%1.1f%%` | 소수점 한 자리 | `25.0%`  |
| `%1.2f%%` | 소수점 두 자리 | `25.00%` |


```python
# 다중 레이아웃 - subplot(행, 열, 순번) 기본

x = np.linspace(0, 2*np.pi, 100)

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title('사인 그래프')

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title('코사인 그래프')

plt.tight_layout()
plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 순번) 미사용

x = np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x))
plt.title('사인 그래프')

plt.plot(x, np.cos(x))
plt.title('코사인 그래프')

plt.tight_layout()
plt.show()
```

```python
# 다중 레이아웃 - subplot(행, 열, 순번) 기본

x = np.linspace(0, 2*np.pi, 100)

for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.plot(x, np.sin(x+i), label=f'sin(x+{i})')
    plt.title(f'{i+1} 사인 그래프')
    #plt.xticks([])
    #plt.yticks([])
    plt.legend()

plt.tight_layout()
plt.show()
```
