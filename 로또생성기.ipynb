```python
import random

# 번호별 출현 횟수(너가 제공한 데이터 그대로 입력)
freq = {
    1:196, 2:185, 3:198, 4:193, 5:175, 6:195,
    7:197, 8:177, 9:156, 10:184, 11:188, 12:203,
    13:201, 14:191, 15:187, 16:192, 17:199, 18:190,
    19:187, 20:195, 21:188, 22:160, 23:164, 24:192,
    25:168, 26:193, 27:202, 28:175, 29:168, 30:185,
    31:191, 32:175, 33:202, 34:204, 35:187, 36:183,
    37:194, 38:194, 39:187, 40:192, 41:162, 42:176,
    43:197, 44:182, 45:190
}

def weighted_lotto():
    numbers = list(freq.keys())
    weights = list(freq.values())

    # 6개 뽑되, 중복 없이 선택
    result = random.choices(numbers, weights=weights, k=20)
    result = list(dict.fromkeys(result))  # 중복 제거
    result = sorted(result[:6])  # 앞에서 6개만 정렬
    return result

# 5게임 추천
for i in range(5):
    print(f"추천 {i+1}: {weighted_lotto()}")
```
