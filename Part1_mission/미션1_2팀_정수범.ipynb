### 문제2. 모음 제거하기

````python
# 문제
def remove_vowels(s):
    vowels = ['a', 'e', 'i', 'o', 'u']
    result = ''
    for char in s:
        if char not in vowels:
            result += char
    return result
````

### 문제6. 짝수 합 구하기

````python
# 문제
def sum_even_numbers(n: int):
    if n == 0:
        print("종료")
        return

    total = 0
    for i in range(1, n + 1):
        if i % 2 == 0:
            total += i

    print(f'1부터 {n}까지의 짝수 합은 {total}입니다.')
````

### 문제1. 문자열 압축 게임



````python
# 문제
def compress_string(s: str) ->str:
    result = ""
    char = s[0]
    count = 1

    for char2 in s[1:]:
        if char2 == char:
            count += 1
        else:
            result += char + str(count)
            char = char2
            count = 1

    result += char + str(count)

    return result
````

````python
# 프로그램 실행
compress_string("aaabbaaa")
````

````python
compress_string("aabcccccaaa")
````

### 문제3. 카운트다운 타이머

````python
# 문제
import time
def countdown_timer(seconds):
    print("타이머 시작!")

    i = 0

    for i in range(seconds, 0, -1):
        print(i)
        time.sleep(1)
        i -= i -1
    print("타이머 종료!")
````

### 문제1. 시간 추적 클래스(TimeTracker)

````python
from datetime import datetime

class TimeTracker:
    def __init__(self):
        self.start_time = None
        self.stop_time = None
        self.elapsed = None

    def start(self):
        self.start_time = datetime.now()
        print("공부 시작")

    def stop(self):
        self.stop_time = datetime.now()
        print("공부 종료")

    def elapsed_time(self):
        if self.start_time and self.stop_time:
            self.elapsed = self.stop_time - self.start_time
            return self.elapsed
        else:
            return
````

````python
# 사용 예시
# 활동 시작
study_session = TimeTracker()
study_session.start()
````

````python
# 활동 종료
study_session.stop()
print("공부한 시간:", study_session.elapsed_time(), "분")
````

### 문제1. 투표 시스템 클래스

````python
# 문제
class VoteSystem:
    def __init__(self):
        self.votes: dict[str, int] = {}

    def add_candidate(self, name):
        name = name.strip()
        if name in self.votes:
            print(f"{name} 후보는 이미 등록되어 있습니다.")
            return
        self.votes[name] = 0
        print(f"{name} 후보가 성공적으로 등록되었습니다.")

    def vote(self, name):
        if name not in self.votes:
            print(f"등록되지 않은 후보입니다: {name}")
            return
        self.votes[name] += 1
        print(f"{name}에게 투표하였습니다.")

    def get_results(self):
        print("투표 결과:")
        if not self.votes:
            print("(등록된 후보가 없습니다.)")
            return
        for candidate, count in self.votes.items():
            print(f"{candidate}: {count} votes")
````

````python
# 사용 예시
voting_system = VoteSystem()
voting_system.add_candidate("Alice")
voting_system.add_candidate("Bob")
voting_system.add_candidate("Charlie")

voting_system.vote("Alice")
voting_system.vote("Alice")
voting_system.vote("Bob")

voting_system.get_results()
````

### 문제2. 은행 계좌 클래스

````python
# 문제
class BankAccount:
    def __init__(self, acount, owner, balance=0):
        self.acount = acount
        self.owner = owner
        self.balance = balance
        print(f"{self.owner}님의 계좌 {self.acount}가 개설되었습니다. 초기 잔액: {balance}")

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount}원 입금되었습니다. 현재 잔액: {self.balance}원")

    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        self.balance -= amount
        print(f"{amount}원 출금되었습니다. 현재 잔액: {self.balance}원")

    def get_balance(self):
        return self.balance
````

````python
# 사용 예시
my_account = BankAccount("123-456-789", "김철수", 100000)
my_account.deposit(50000)
my_account.withdraw(20000)
print(f"현재 잔액: {my_account.get_balance()}원")
````

### 문제3. 직원 관리 클래스

````python
# 문제
class EmployeeManager:
    employees: dict[str, int] = {}

    def __init__(self, name=None, salary=None):
        if name and salary:
            EmployeeManager.employees[name] = salary
            print(f"{name} 님이 추가되었습니다. 급여: {salary}원")

    @classmethod
    def calculate_average_salary(cls) -> None:
        if not cls.employees:
            print("직원이 없습니다.")
            return
        avg = sum(cls.employees.values()) / len(cls.employees)
        print(f"전체 직원 평균 급여: {avg}원")
````

````python
# 사용 예시
emp1 = EmployeeManager("홍길동", 50000)
emp2 = EmployeeManager("김철수", 60000)

EmployeeManager.calculate_average_salary()
````

### 문제4. 프랜차이즈 레스토랑 관리 클래스

````python
# 문제
class ReservationSystem:
    def __init__(self, branch_name):
        self.branch_name = branch_name
        self.reservations = []

    def add_reservation(self, name, date_str, people):
        self.reservations.append((name, date_str, people))
        return

    def list_reservations(self):
        print(f"{self.branch_name} 예약 목록:")
        if not self.reservations:
            print("- (예약 없음)")
            return
        for name, date_str, people in self.reservations:
            print(f"- {name}, {date_str}, {people}명")

    @classmethod
    def sum_reservations(cls, systems):
        total = 0
        for sys in systems:
            for name, date_str, people in sys.reservations:
                total += people
        return total
````

````python
# 사용 예시
restaurant1 = ReservationSystem("강남점")
restaurant2 = ReservationSystem("홍대점")

restaurant1.add_reservation("홍길동", "2024-05-20", 4)
restaurant2.add_reservation("김철수", "2024-05-21", 2)

restaurant1.list_reservations()
restaurant2.list_reservations()

total_reservations = ReservationSystem.sum_reservations([restaurant1, restaurant2])
print(f"전체 레스토랑 예약 수: {total_reservations}")
````

## 주요 개선점

### 1. **Return vs Print 명확히 구분**
- 함수에서 값을 반환할 때는 `return` 사용
- 화면에 출력할 때는 `print` 사용
- 둘을 혼동하면 함수가 `None`을 반환하게 됩니다

### 2. **문자열 처리 주의**
- 문자열을 반환할 때 불필요한 따옴표를 추가하지 마세요
- 출력 형식을 정확히 맞춰주세요

### 3. **클래스 변수 vs 딕셔너리**
- 직원 관리에서는 직원 객체들을 리스트로 저장해야 합니다
- 딕셔너리로 저장하면 객체의 다른 정보들이 손실됩니다
